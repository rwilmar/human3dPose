```python
import cv2
import numpy as np

import pandas as pd

from inference import Network
from model_handlers import handle_output
from image_handlers import preprocessing, create_output_image

from matplotlib import pyplot as plt
```

```python
MODEL="./models/single-human-pose-estimation-0001.xml" 
CPU_EXT=None
DEVICE="CPU"
ModelId="SINGLEPOSE"


image_1="./capt.png" 
```

```python
# Create a Network for using the Inference Engine
inference_network = Network()
    
# Load the model in the network, Read network parameters (shape)
n, c, h, w = inference_network.load_model(MODEL, DEVICE, CPU_EXT)

# Read the input image
image = cv2.imread(image_1)
preprocessed_image = preprocessing(image, h, w)
```

```python
# Perform synchronous inference on the image
inference_network.sync_inference(preprocessed_image)
# Obtain the output of the inference request
output = inference_network.extract_output()
```

```python
processed_output = handle_output(ModelId)(output, image.shape)
#try .... 
output_image = create_output_image(ModelId, image, processed_output)
cv2.imwrite("./output_net.png", output_image)
```

```python

#im_rgb = cv2.cvtColor(output_image, cv2.COLOR_RGB2BGR)
plt.imshow(output_image)
```

```python
#dibujar articulacion
art=9

artLayer=processed_output[art]
print(artLayer.min(), artLayer.max())
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(artLayer)
print(max_loc)

imout= np.zeros(image.shape)
imout[:,:,1]=artLayer
cv2.circle(imout, max_loc, 3, (0, 200, 200), 5)

imout=imout.astype(int)+cv2.cvtColor(image, cv2.COLOR_BGR2RGB)#image.astype(int)
plt.imshow(imout)
```

```python
# dibujar skeleton

joints, height, width = processed_output.shape

# initialize dataframe of body parts
ixn=['head','left_eye','right_eye','left_ear','right_ear','left_shoulder','right_shoulder','left_elbow',
    'right_elbow','left_hand', 'right_hand','left_hip', 'right_hip','left_knee','right_knee', 
    'left_foot', 'right_foot' ]
max1d=pd.DataFrame(np.zeros((17,2), dtype=np.intc), index=ixn, columns=['x_max','y_max'])

for bodyPart in range(joints):
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(processed_output[bodyPart])
    max1d.iloc[bodyPart, 0:2] = max_loc

max1d
```

```python

imout= np.zeros(image.shape, dtype=np.int8)
#imout.shape

cv2.line(imout, tuple(max1d.loc['head']), tuple(max1d.loc['right_eye']), (0, 200, 200),5)
cv2.line(imout, tuple(max1d.loc['head']), tuple(max1d.loc['left_eye']), (0, 200, 200),5)
cv2.line(imout, tuple(max1d.loc['head']), tuple(max1d.loc['right_hip']), (0, 200, 200),5)
cv2.line(imout, tuple(max1d.loc['head']), tuple(max1d.loc['left_hip']), (0, 200, 200),5)

cv2.line(imout, tuple(max1d.loc['left_shoulder']), tuple(max1d.loc['right_shoulder']), (0, 200, 200),5)


cv2.line(imout, tuple(max1d.loc['left_shoulder']), tuple(max1d.loc['left_elbow']), (0, 200, 200),5)
cv2.line(imout, tuple(max1d.loc['left_elbow']), tuple(max1d.loc['left_hand']), (0, 200, 200),5)
cv2.line(imout, tuple(max1d.loc['right_shoulder']), tuple(max1d.loc['right_elbow']), (0, 200, 200),5)
cv2.line(imout, tuple(max1d.loc['right_elbow']), tuple(max1d.loc['right_hand']), (0, 200, 200),5)

cv2.line(imout, tuple(max1d.loc['left_hip']), tuple(max1d.loc['right_hip']), (0, 200, 200),5)

cv2.line(imout, tuple(max1d.loc['right_hip']), tuple(max1d.loc['right_knee']), (0, 200, 200),5)
cv2.line(imout, tuple(max1d.loc['right_knee']), tuple(max1d.loc['right_foot']), (0, 200, 200),5)
cv2.line(imout, tuple(max1d.loc['left_hip']), tuple(max1d.loc['left_knee']), (0, 200, 200),5)
cv2.line(imout, tuple(max1d.loc['left_knee']), tuple(max1d.loc['left_foot']), (0, 200, 200),5)


plt.imshow(imout)
```
